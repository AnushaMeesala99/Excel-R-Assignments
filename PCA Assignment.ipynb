```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("wine.csv")
data 
```

```python
wine= data.iloc[:,1:]
wine 
```

```python
wine.values
```

```python
# Normalizing the numerical data 

wine_normal = scale(wine)
wine_normal
```

```python
#Transforming to principal components
pca = PCA(n_components=13)
pca_values=pca.fit_transform(wine_normal)
pca_values 
```

```python
pca.components_
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var 
```

```python
var1 = np.cumsum(var*100)
var1 
```

```python
# Variance plot for PCA components obtained 

plt.plot(var1,color="red")
```

```python
pca_values[:,[0,1,2]]
```

```python
data1=pd.DataFrame( pca_values[:,[0,1,2]])
data2 = data1.rename(columns={0:"pc1",1:'pc2',2:'pc3'})
data2
```

```python
wine_data= data['Type']
wine_data
data2['Type']= wine_data 
data2 
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.scatterplot(data=data2) 
```

# Checking with hierarchial clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
from scipy.cluster.hierarchy import linkage 
```

```python
z= linkage(wine_normal, method="complete",metric= "euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete=AgglomerativeClustering(n_clusters=3, linkage="complete", affinity="euclidean").fit(wine_normal)
clusterlabels = pd.Series(h_complete.labels_)
clusterlabels
data["Clusters"]=clusterlabels  # creating a  new column and assigning it to new column 
data
```

```python
data.Clusters.value_counts()
```

# Checking with KMEANS clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,8):
    clf= KMeans(n_clusters=i)
    clf.fit(wine_normal)
    wcss.append(clf.inertia_)
plt.plot(range(1,8),wcss)
plt.ylabel("WCSS") 
plt.xlabel("Number of Clusters")
plt.title("The Elbow Curve")
```

```python
from sklearn.cluster import KMeans
clf=KMeans(n_clusters=3) 
y_kmeans = clf.fit_predict(wine_normal)
y_kmeans
```

```python
clf.labels_
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
data['clusters']=md # creating a  new column and assigning it to new column 
data
```
