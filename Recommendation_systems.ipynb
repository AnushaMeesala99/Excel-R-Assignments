```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("books.csv")
data1 =data.drop(columns='Unnamed: 0')
data1
```

```python
books= data1.rename(columns={'User.ID':'User_ID','Book.Title':'Book_Title','Book.Rating':'Book_Rating'})
```

```python
books
```

```python
import seaborn as sns
sns.pairplot(books)
```

```python
len(books.User_ID.unique())
```

```python
len(books.Book_Title.unique())
```

```python
#Checking for duplicates
books.duplicated(subset=['User_ID', 'Book_Title']).sum()
```

```python
books.drop_duplicates(subset=["User_ID","Book_Title" ],inplace=True)
books
```

```python
user_books_df = books.pivot(index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating').reset_index(drop=True)
```

```python
user_books_df
```

```python
user_books_df.index = books.User_ID.unique()
```

```python
user_books_df
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
user_books_df 
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_books_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books.User_ID.unique()
user_sim_df.columns = books.User_ID.unique()
user_sim_df
```

```python
user_sim_df.iloc[0:20, 0:20]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax()[0:50]
```

```python
#IBCF
#Fetching ratings for Decision in Normandy
Cannibal_In_Manhattan = user_books_df['Cannibal In Manhattan']  
```

```python
Cannibal_In_Manhattan
```

```python
#Finding the correlation with different movies
similar_to_Cannibal = user_books_df.corrwith(Cannibal_In_Manhattan) 
```

```python
similar_to_Cannibal.head(10)
```

```python
corr_cannibal = pd.DataFrame(similar_to_Cannibal, columns=['Correlation'])
corr_cannibal.dropna(inplace=True)
corr_cannibal.head()
```

```python
corr_cannibal[corr_cannibal['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(20)  
```
