# Hierarchial clustering

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("crime_data.csv")
data
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(data.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=4, linkage='complete',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
data['clust']=cluster_labels # creating a  new column and assigning it to new column 
data
```

```python
data.iloc[:,1:].groupby(data.clust).mean()
```

# KMEANS CLUSTERING

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
model = KMeans(n_clusters=4)
```

```python
y_kmeans = model.fit_predict(df_norm)
```

```python
model.labels_
```

```python
kdata = pd.read_csv("crime_data.csv")
kdata
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
kdata['clust']=md # creating a  new column and assigning it to new column 
kdata
```

```python
kdata.iloc[:,1:5].groupby(kdata.clust).mean()
```

```python
kdata.plot(x="Murder",y ="Assault",c=model.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```
